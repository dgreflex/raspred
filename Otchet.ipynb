# Итог

Для транспьютерной матрицы 5*5 с $T_s = 100$ и $T_b = 1$ время передачи составляет $\approx 1.8343 * L $ где  $ L $ - стартовый размер сообщения

$t_{итог} = T_s + 1.8333 * L * T_b$

Асимптотическая сложность O(L)
